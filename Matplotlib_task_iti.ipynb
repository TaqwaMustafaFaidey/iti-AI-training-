```python
df_line = pd.DataFrame({'X': np.linspace(0, 10, 100), 'Y': np.linspace(0, 10, 100)/2})
df_scatter = pd.DataFrame({'X': np.random.rand(10), 'Y': np.random.rand(10)})
drinks = {"Coffee": 120, "Tea": 75, "Iced Coffee": 90}
df_bar = pd.DataFrame(list(drinks.items()), columns=['Drink', 'Price'])
df_hist = pd.DataFrame({'Value': np.random.randn(1000)})
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 6))
axes[0, 0].plot(df_line['X'], df_line['Y'])
axes[0, 1].scatter(df_scatter['X'], df_scatter['Y'])
axes[1, 0].bar(df_bar['Drink'], df_bar['Price'])
axes[1, 1].hist(df_hist['Value'], bins=30)
plt.tight_layout()
plt.show()
```

```python
car_sales = pd.read_csv('/car-sales - car-sales.csv')
car_sales['Price'] = car_sales['Price'].replace('[\$,]', '', regex=True).astype(float)
df_price = pd.DataFrame({'Price': car_sales['Price']})
df_price.plot()
plt.show()
```

```python
from datetime import datetime
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales['Sale Date'] = pd.date_range(start=datetime.today(), periods=len(car_sales), freq='D')
df_sales = pd.DataFrame({'Sale Date': car_sales['Sale Date'], 'Total Sales': car_sales['Total Sales'], 'Price': car_sales['Price'], 'Odometer': car_sales['Odometer (KM)']})
df_sales.plot(x='Sale Date', y='Total Sales')
plt.show()
df_sales.plot(x='Odometer', y='Price', kind='scatter')
plt.show()
```
